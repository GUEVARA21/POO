````python
import matplotlib.pyplot as plt #se importa de la libreria plt

class Punto(): #Se crea una clase llamada punto
    def __init__(self, x, y): #Se definen las variables dentro de nuestra clase
        self.x = x #Estas es una variable
        self.y = y #Esta es otra variable

    def plot(self): #Se crea otra función.
        plt.scatter(self.x, self.y, color='red', marker='o') #Cararterizcas del plano cartesiano
        plt.xlabel('X-axis') #Le da nombre al eje de las abcisas
        plt.ylabel('Y-axis') #Le da nombre al eje de las ordenadas
        plt.title('Gráfica del punto') #Es el nombre de la grafica
        plt.grid(True)
        plt.show() #Muestra  a la grafica
````

````python
Punto(4,6).plot() #Se usa el metodo-plotpoint para grafica
````

````python
class aviones():
    def __init__(self, aerolinea, gasolina):
        self.aerolinea = aerolinea
        self.gasolina = gasolina

    def destino(self):
        print('!Donde quieres ir!')

class jumbo(aviones):
    def __init__(self, aerolinea, gasolina, tamaño):
        super().__init__(aerolinea, gasolina)
        self.tamaño = tamaño

    def destino(self, llave):
        if llave == 'llave_avion':
            print('¡Listo para irnos de viaje!')
        else:
            print('La llave no es la correcta')

    def apagar(self, apagar):
        if apagar:
            print('¡El avion no tiene combustible!')
        else:
            print('¡Estamos en mediovuelo')
````
